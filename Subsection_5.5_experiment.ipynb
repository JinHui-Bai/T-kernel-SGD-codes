```python
import numpy as np
class T_SGD():
    def __init__(self,s,r,theta,d,gamma_0):
        self.L_n = 1
        self.s = s
        self.r = r
        self.theta = theta
        self.d = d
        self.gamma0 = gamma0
        self.fn = np.zeros(1)
        self.fn_ave = np.zeros(1)
        self.f_rho = np.array([0]+[ (j+1)**(-2*self.s*self.r-0.5) for j in range(self.d)])

    def comp_f_rho(self,X):
        return np.dot(self.f_rho,X)

    def comp_f_n(self,X):
        return np.dot(self.fn,X[:self.L_n])

    def comp_f_ave(self,X):
        return np.dot(self.fn_ave,X[:self.L_n])
  
    def construct_data(self):
        X = np.random.randn(self.d+1)
        for i in range(1,self.d+1):
            X[i] = X[i]*(i**(-self.s))
        Y = self.comp_f_rho(X)
        Z = Y + np.random.normal(0,0.5)
        return [X,Z,Y]
    
    def updata_T_kernel_SGD(self,X,Y):
        if self.L_n < self.epoch**(self.theta) +1:
            self.L_n += 1 
            A = np.zeros(1)
            self.fn = np.append(self.fn,A.copy())
            self.fn_ave = np.append(self.fn_ave,A.copy())
        coe = self.gamma0*(Y-self.comp_f_n(X))
        self.fn = self.fn + coe*X[:self.L_n]
        self.fn_ave = (self.epoch*self.fn_ave+self.fn)/(self.epoch+1)
        
    def test(self):
        S1 = 0
        S2 = 0
        for i in range(1000):
            X,Z,Y = self.construct_data()
            S1 += (Y-self.comp_f_n(X))**2
            S2 += (Y-self.comp_f_ave(X))**2
        return [S1/1000,S2/1000]
    
    def  algorithm_sgd(self,sum_epoch,NN):
        ERR_fn = []
        ERR_fn_ave = []
        self.epoch = 1
        for i in range(sum_epoch):
            X,Z,Y = self.construct_data()
            self.updata_T_kernel_SGD(X,Z)
            if self.epoch in NN:
                err_fn,err_fn_ave = self.test()
                print(self.epoch, err_fn,err_fn_ave)
                ERR_fn.append(err_fn)
                ERR_fn_ave.append(err_fn_ave)
            self.epoch += 1
        #print(ERR_fn)
        print(ERR_fn_ave)
        print(self.L_n)
        return ERR_fn_ave
```

```python
s = 1.
r = 0.7
theta = 1/(1+4*s*r)
d = 1000
gamma0 = 0.2
T = 400001
t_sgd = T_SGD(s,r,theta,d,gamma0)
N = [int(10**(i/5))for i in range(1,40)]
t_sgd_err = t_sgd.algorithm_sgd(T,N)
```
